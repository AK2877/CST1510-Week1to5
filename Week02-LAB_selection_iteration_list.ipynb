4. **Body mass index (BMI)** is a measure of health and weight.

It can be calculated by taking your weight in kilograms and dividing it by the square of your height in meters. The interpretation of BMI for people 20 years or older is as follows:


BMI	Interpretation
BMI < 18.5	Underweight
18.5 ≤ BMI ≤ 25.0	Normal
25.0 ≤ BMI ≤ 30.0	Overweight
30.0 ≤ BMI 	Obese
 
Write a program that prompts the user to enter weight in pounds and height in inches and display the BMI. Note that one pound is 0.45359237 kilograms, and one inch is 0.0254 meters


````python
w = float(input("Enter weight in pounds: "))
h = float(input("Enter height in inches: "))

w *= 0.45359237
h *= 0.0254

bmi = w / (h*h)
print(f"\nYour BMI is: {bmi:.1f}")

if bmi < 18.5:
    print("Underweight")
elif bmi >= 18.5 and bmi <= 25.0:
    print("Normal")
elif bmi >= 25.0 and bmi <= 30.0:
    print("Overweight")
elif bmi <= 30.0:
    print("Obese")
````

5. **Student Marks Analysis**
You are working as a data analyst in a school. You have the exam marks of students stored in a list. Your task is to analyze the data.

**Requirements**:

- Store marks of students in a list.

- Use a loop to calculate the average mark.

- Find the highest and lowest mark using list functions.

- Print the number of students who passed (marks ≥ 50) and failed.

````python
marks = [30,65,95,24,66,89,69,77]

low = min(marks)
high = max(marks)

cnt_p = cnt_f = 0
avg = 0

for i in marks:
    avg += i
    if i >= 50:
        cnt_p += 1
    else:
        cnt_f += 1
        
avg /= len(marks)

print(f" Highest marks = {high}\n Lowest marks = {low}\n Average marks = {avg:.2f}")
print(f" Number of Students who passed: {cnt_p}\n Number of Students who failed: {cnt_f}")
````

18. Suppose you want to develop a **program to play a lottery.**
    
The program randomly generates a two-digit number, prompts the user to enter a <u>two-digit number</u>, and determines whether the user wins according to the following rules:


    1. If the user’s input matches the lottery in the exact order, the award is £10,000.
    2. If all the digits in the user’s input match all the digits in the lottery number, the award is £3,000.
    3. If one digit in the user’s input matches a digit in the lottery number, the award is £1,000.



````python
import random as r

rand_no = r.randint(10,99)  # 2-digit random number between 10 & 99
n = int(input("input your 2-digit Lottery Number: "))

print(f"\nYour number: {n}\nLottery Number: {rand_no}")

# checks for invalid inputs
if n < 10 or n > 99:
    print ("invalid Number")
    
elif rand_no == n:
    print("Congratulations! You have Won the Lottery!! Award: £10,000")
        
else:
    rev = 0
    m = n
    
    # loop to get the reverse of the user defined number
    while m > 0:
        temp = m % 10
        rev = (rev * 10) + temp
        m //= 10
        
    if rand_no == rev:
        print("You have been awarded £3,000")

    else:
        # converts the 2 numbers into string & checks if any of the digits match
        str_rand_no = str(rand_no)
        str_n = str(n)
        cnt = 0
        
        # loop to find the number of matching digits
        for i in range (0,2):
         for j in range (0,2):
            if str_rand_no[i] == str_n [j]:
                cnt += 1
                
        if cnt >= 1:
            print("You have been awarded £1,000")
        else:
            print("Better luck next time")   
````

20. (Find the two highest scores) - Write a program that prompts the user to enter the number of students and each student’s score and displays the highest and second highest scores.


````python
# Amaan Ali Khan, M01095853
n = int(input("Enter the number of students: "))
high = sec_high = -1  # initialize the highest and second highest number

# loop to get the scores of each student
for i in range(n):
    score = float(input(f"Enter the score for student {i + 1}: "))

    # checks if the current score is greater than the highest
    if score > high:
        sec_high = high
        high = score
        
    # checks if the current score is greater than the second highest    
    elif score > sec_high and score < high:
        sec_high = score
        
print(f"Highest score: {high:.2f}")
print(f"Second highest score: {sec_high:.2f}")
````
